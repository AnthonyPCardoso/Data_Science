```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataset.csv")
```

```python
sns.histplot(data=df['horas_estudo_mes'], color='red', kde=True)
```

```python
x = np.array(df['horas_estudo_mes'])
```

```python
x = x.reshape(-1, 1)
```

```python
y = df['salario']
```

```python
plt.scatter(x,y, color='red', label='Disperção')
plt.xlabel('Horas de estudos')
plt.ylabel('Salário')
plt.legend()
plt.show()
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x_treino,y_treino)
```

```python
plt.scatter(x, y, color='blue', label='Dados Reais Históricos')
plt.plot(x, modelo.predict(x), color='black', label='Reta de regressão Linear')
plt.xlabel('Horas trabalhando')
plt.ylabel('Salário')
plt.legend()
plt.show()
```

```python
score = modelo.score(x_teste, y_teste)
print(f"Coeficiente: {score:.2f}")
```
